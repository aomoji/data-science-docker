```python
def plot_time(df):
    data = pd.concat(
        [
            df.groupby("受注日")["総合計"].sum(),
            df.groupby(["受注日", "購入者名"])
            .count()
            .groupby("受注日")[["発送方法"]]
            .count(),
        ],
        axis=1,
    )
    g1 = sns.lineplot(data=data["総合計"], color="g")
    plt.xticks(rotation=40)
    # g1.legend("a")
    ax2 = plt.twinx()
    g2 = sns.lineplot(data=data["発送方法"], color="b", ax=ax2)  # , label="")
    # g2.legend("b")
    plt.show()
```

```python
for year in range(2014,2019):
    plt.figure(figsize=(20, 10))
    y_data = data[data.year == year]
    plot_time(y_data)
```
